```python
import parlai
print(parlai.__version__)
```

```python
from parlai.core.params import ParlaiParser
from parlai.core.agents import create_agent
```

```python
parser = ParlaiParser(True, True)
opt = parser.parse_args(
    args = [
        '--model-file', 'zoo:blenderbot2/blenderbot2_3B/model',
        '--query-generator-model-file', 'zoo:sea/bart_sq_gen/model',
        #'--rag-retriever-type', 'search_term_faiss',
        '--search-server', '0.0.0.0:8889',
        '--memory-key', 'full_text',
        '--knowledge-access-method', 'all',
    ]
)
```

```python
agent = create_agent(opt)
```

```python
agent.opt.log()
```

```python
agent.history.reset()
```

```python
agent.observe({
    'text': 'your persona: i am richard lionheart.\nyour persona: i am the king of england.',
    'episode_done': False,
})

agent.observe({
    'text': 'do you have any hobbies?',
    'episode_done': False,
})
```

```python
agent.act()
```

```python
agents = [
    agent,
    agent.clone(),
    agent.clone(),
]
```

```python
def reset_agent(agent):
    agent.history.reset()
    agent.observe({
        'text': 'your persona: i am richard lionheart.\nyour persona: i am the king of england.',
        'episode_done': False,
    })
```

```python
for a in agents:
    reset_agent(a)
```

```python
agent_pool = {
    a for a in agents
}
```

```python
assigned_agents = {}
```

```python
import eventlet
import socketio

sio = socketio.Server()
app = socketio.WSGIApp(sio, static_files={
    '/': {'content_type': 'text/html', 'filename': 'static/index.html'},
    '/css': './static/css',
    '/img': './static/img',
    '/js': './static/js', 
})

@sio.event
def connect(sid, environ):
    print('connect ', sid)
    if agent_pool:
        assigned_agents[sid] = agent_pool.pop()

@sio.event
def message(sid, data):
    print('message ', data)
    if sid in assigned_agents:
        agent = assigned_agents[sid]
        agent.observe({
            'text': data['data'],
            'episode_done': False,
        })
        obs = agent.act()
        print('sending response', obs['text'])
        sio.emit('response', {'data': obs['text']})

@sio.event
def disconnect(sid):
    print('disconnect ', sid)
    if sid in assigned_agents:
        agent = assigned_agents.pop(sid)
        reset_agent(agent)
        agent_pool.add(agent)

eventlet.wsgi.server(eventlet.listen(('', 5000)), app)
```
